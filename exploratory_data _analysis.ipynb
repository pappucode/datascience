### Exploratory Data analysis
#### The purpose of EDA

```python
#Investigate about the dataset and variables
#Investigate about the null value and other unwanted value
#Know the relationship betweeen different relationship
#Know the descriptive statistics of the dataset
#To extract the hidden information
#Easier the decistion taking, when need to fit the machine learning model
```

#### Load the dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab

df = pd.read_csv('data-dhaka-weather1953-2016.csv')
df.head()
```

#### Check the null value existence

```python
df.isnull().values.any()
```

#### Descriptive Statistics

```python
df.describe()
```

#### Correlation

```python
df.corr()
```

#### Heatmap

```python
correlation = df.corr()
plt.figure(figsize=(16,8))
sns.heatmap(correlation, annot=True, linewidths=1, vmin=-1, cmap="RdBu_r")
```

#### Find the monthly mean Rainfall

```python
rain = df.groupby('Month')['Rainfall'].mean()
rain
```

```python
#So here we understand that rainfall is usually higher in the middle of the year. On the other hand, rainfall is relatively low at the beginning and end of the year.
```

#### Distribution Plot

```python
sns.distplot(rain)
```

#### The wettest month

```python
df.groupby('Month')['Rainfall'].mean().sort_values(ascending=False)
```

#### Boxplot of the whole dataset

```python
plt.figure(figsize=(16, 6))
ax = sns.boxplot(data=df, orient="h", palette="Set2")
```

#### Separate boxplot

```python
ax = sns.boxplot(x=df["Rainfall"])
```

#### Boxplot of two variables together

```python
plt.figure(figsize=(16, 6))
ax = sns.boxplot(x="RelativeHumidity", y='Rainfall', data=df)
```

```python
#Here we see that with the increase in humidity, the rainfall has also increased.
```

#### Category Box Plot

```python
plt.figure(figsize=(16, 6))
ax = sns.catplot(x='Month', y="Rainfall", kind="box", data=df)
```

```python
#Boxplotting the rainfall by month we see that the amount of rainfall is higher in the middle months of the year, the data spread of these months is also higher. Again, the data spread of the months in which the rainfall is less is less.
```

```python
df.hist()
```

#### Scatter plot

```python
#Scatter plot between Humidity and Rainfall
plt.scatter(df.Rainfall, df.RelativeHumidity)
```

```python
#Here the scatter plot show that when the humidity is incresed, the rainfall also increased
```

#### Join Plot


```python
#It is mainly used to visualize scatter plots and distribution plots together.
sns.jointplot(x="Rainfall", y="RelativeHumidity", data=df, kind="reg");
```
