### Association Rule 

##### Intoduction of dataset and load packages

We'll use a Walmart dataset to find out the association rules. This dataset contains the transaction data of Walmart chain shops in different countries.

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

df = pd.read_excel("Online_Retail.xlsx")
df.head()
```

##### See the counties in this dataset

```python
df.Country.unique()
```

#### Clean the dataset

```python
#Stripping extra spaces in the description
df['description'] = df['Description'].str.strip()

# Dropping the rows without any invoice Number
df.dropna(axis=0, subset=['InvoiceNo'], inplace=True) 
df['InvoiceNo'] = df['InvoiceNo'].astype('str')

#Dropping all transactions which ware done on credit
df = df[~df['InvoiceNo'].str.contains('C')]
```

#### This time we will separate data from dataset only for Italy

```python
basket_Italy = (df[df['Country'] =="Italy"] 
          .groupby(['InvoiceNo', 'Description'])['Quantity'] 
          .sum().unstack().reset_index().fillna(0) 
          .set_index('InvoiceNo'))
basket_Italy
```

```python
def hot_encode(x):
    if x <= 0:
        return 0
    if x >= 0:
        return 1
    
basket_encoded = basket_Italy.applymap(hot_encode)
basket_Italy = basket_encoded
basket_Italy
```

#### Now create the Association rules. We will make rules with a minimum support value of 0.05 or 5%

```python
fre_items = apriori(basket_Italy, min_support=0.05, use_colnames=True)

#Collect the inferrd rules in a dataframe
rules = association_rules(fre_items, metric="lift", min_threshold=1)
rules = rules.sort_values(['confidence', 'lift'], ascending=[False, False])
rules
```
