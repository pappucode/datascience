### Two options for subplots 

```python
#Option 1
fig, ((ax1, ax2), (ax3,ax4)) = plt.subplots(ncols=2, nrows=2, figsize=(10,5))

ax1.plot(x,x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
#Option 2
fig, ax = plt.subplots(ncols=2, nrows=2, figsize=(10,5))
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

#### pyplot vs matplotlib OO method
* When plotting something quickly,okay to use the pyplot method
* When plotting something more advanced, use the OO method


```python
heart_disease.head()
```

```python
age_over_50 = heart_disease[heart_disease['age']>50]
age_over_50.head()
```

```python
# pyplot method
age_over_50.plot(kind='scatter', x='age', y='chol', c='target', figsize=(10,7));
```

```python
# OO METHOD mixed with pyplot method
fig, ax = plt.subplots(figsize=(10,7))
age_over_50.plot(x='age', y='chol', c='target', ax=ax, kind='scatter');
ax.set_xlim([45,100])
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10,7))

# plot the data
scatter = ax.scatter(x=age_over_50['age'], y=age_over_50['chol'], c=age_over_50['target'])

#Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels", xlabel="Age", ylabel="Cholesterol");

#Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

#add a horizontal line
ax.axhline(y=age_over_50['chol'].mean(), linestyle="--");
```

```python
# Subplot of chol, age and thalach
fig, (ax0, ax1) = plt.subplots(ncols=1,nrows=2, figsize=(10,10), sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=age_over_50['age'], y=age_over_50['chol'], c=age_over_50['target'])

#customize ax0
ax0.set(title="Heart disease and cholesterol level", ylabel="Cholesterol");

#add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target");

#add a horizontal line
ax0.axhline(y = age_over_50['chol'].mean(), linestyle="--");


#add data to ax1
scatter = ax1.scatter(x=age_over_50['age'], y=age_over_50['thalach'], c=age_over_50['target'])

#customize ax1
ax1.set(title="Heart desease and max heart rate", xlabel="Age", ylabel="Max heart rate")

#add legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target");

#add horizontal line to ax1
ax1.axhline(y=age_over_50['thalach'].mean(), linestyle="--");

#Add a title for the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
#set the style
plt.style.use('seaborn-whitegrid')
# OO method from scratch
fig, ax = plt.subplots(figsize=(10,7))

# plot the data
scatter = ax.scatter(x=age_over_50['age'], y=age_over_50['chol'], c=age_over_50['target'], cmap='winter')

#Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels", xlabel="Age", ylabel="Cholesterol");

#Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

#add a horizontal line
ax.axhline(y=age_over_50['chol'].mean(), linestyle="--");
```

```python
#Customize the y and x axis limitations

# Subplot of chol, age and thalach
fig, (ax0, ax1) = plt.subplots(ncols=1,nrows=2, figsize=(10,10), sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=age_over_50['age'], y=age_over_50['chol'], c=age_over_50['target'], cmap='winter')

#customize ax0
ax0.set(title="Heart disease and cholesterol level", ylabel="Cholesterol");

#Change the ax0 x axis limits
ax0.set_xlim([50, 80])

#add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target");

#add a horizontal line
ax0.axhline(y = age_over_50['chol'].mean(), linestyle="--");


#add data to ax1
scatter = ax1.scatter(x=age_over_50['age'], y=age_over_50['thalach'], c=age_over_50['target'], cmap='winter')

#customize ax1
ax1.set(title="Heart desease and max heart rate", xlabel="Age", ylabel="Max heart rate")

#Change the ax1 x axis limits
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])

#add legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target");

#add horizontal line to ax1
ax1.axhline(y=age_over_50['thalach'].mean(), linestyle="--");

#Add a title for the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
